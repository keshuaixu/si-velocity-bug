```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('PSM3.csv', delimiter=',')
data.shape
data = data[4200:4700, :]
```

```python
pos = data[:, 0]
dt = data[:, 1]
vel_software = data[:,2]
dpos = np.diff(pos)
dpos = np.hstack([[np.nan], dpos])
vel_my = dpos/dt
# dpos.shape
```

```python
# plt.figure()
fig, ax = plt.subplots(5,1,figsize=(20,10))
ax[0].plot(pos)
ax[1].plot(dt)
ax[2].plot(vel_software)
ax[3].plot(dpos)
ax[4].plot(vel_my)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax2 = ax.twinx()
ax.plot(dt[400:],'.-', alpha=0.7, label='dt')
ax2.plot(dpos[400:], '.-', c='orange', alpha=0.7, label='dpos')
fig.legend()
```
